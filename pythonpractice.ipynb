```python
# This import allows the code to mess around with things on the computer, in our case, files
import os, shutil
# Having the path tells the program where its going to be working. 
# An additional / is included to tell the program to work within the folder specified.
path=r"C:/Users/Miles/Downloads/python file sorting/"
# These two variables set up some prerequisites for our program.
# The first lists the files within the specified path. The second lists the names of folders I would like.
file_name=os.listdir(path)
folder_names = ["excelfiles", "imagefiles", "pdfs"]
# This for loop first checks the number of folders in the folder_names list, then makes said folders if they do not exist within the specified directory
for i in range(len(folder_names)):
    if not os.path.exists(path + folder_names[i]):
        os.makedirs(path + folder_names[i])
# This for loop checks each file in the folder then moves it to the appropriate folder if it does not already exist in the folder. 
for file in file_name:
    if ".xlsx" in file and not os.path.exists(path + "excelfiles" + file):
        shutil.move(path + file, path + "excelfiles/" + file)
    elif ".pdf" in file and not os.path.exists(path + "pdfs" + file):
        shutil.move(path + file, path + "pdfs/" + file)
    elif ".png" in file and not os.path.exists(path + "imagefiles" + file):
        shutil.move(path + file, path + "imagefiles/" + file)
    elif ".jpg" in file and not os.path.exists(path + "imagefiles" + file):
        shutil.move(path + file, path + "imagefiles/" + file)
    
```

```python
from bs4 import BeautifulSoup
import requests
url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue'
#Wikipedia requires that any web scraping programs assign a useragent. The headers variable is used to tell Wikipedia who I am
headers = {'User-Agent': 'PracticeWebScraper (https://github.com/Vonn-K)'}
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html')
# Here we are trying to find the tables of wikipedia so we can eventually put them in a dataframe. 
table = soup.find_all('table')[0]
soup.find('table', class_ = 'wikitable sortable')
```

```python
world_titles = table.find_all('th')
```

```python
world_titles
```

```python
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=world_table_titles)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
df.to_csv(r'C:/Users/Miles/Downloads/Github files/Folder for Output/Companies.csv', index=False)
```

```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.scrapethissite.com/pages/forms/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
table = soup.find('table')
```

```python
table_titles = [title.text.strip() for title in soup.find_all('th')]
print(table_titles)
```

```python
import pandas as pd
df = pd.DataFrame(columns=table_titles)
df
```

```python
row_data = table.find_all('tr')
```

```python
for row in row_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    length = len(df)
    df.loc[length] = individual_row_data
df.to_csv(r'C:/Users/Miles/Downloads/Github files/Folder for Output/HockeyTeams.csv', index=False)
```
