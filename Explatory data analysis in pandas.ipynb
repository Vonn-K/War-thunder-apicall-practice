```python
# Using seaborn, we can build a heat map. this is a way to visualize correlation between items in your data. 
# by using annotation set to true, we get titles. 
sns.heatmap(df.corr(numeric_only=True), annot = True)
# params configures the size of your heatmap. adjust as needed.
plt.rcParams['figure.figsize'] = (30,7)
plt.show()
```

```python
# here, we use the groupby function to group everything by the continent column, then by using an index to get all the items
# that we want, in this case each decades population, we can feed it into a mean function, followed by sorting by descending. 
# we include [::-1] to reverse the order that the groupby function iterates through the columns, in this case
# its because the years are descending from left to right in the original dataframe.
df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
df2
```

```python
# here we transpose our dataframe to swap the columns and rows for a better graph. 
df3 = df2.transpose()
df3
```

```python
df3.plot()
```

```python
# boxplots are very good at finding outliers in your data. the line on the bottom is the average, and the dots represent outliers.
df.boxplot(figsize=(30, 10))
```
