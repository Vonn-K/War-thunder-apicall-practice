```python
specific_companies = ['Walmart', 'Alphabet']
#the isin() function is used to check if a value is in a list, it also works on tuples and sets. It returns a boolean, so you need to put it in
# a second set of brackets to return the rows that return true. 
df[df['Name'].isin(specific_companies)]
```

```python
#str.contains() is used to check if a string is in a column, once again it returns a boolean, so you need to put it in the second set of brackts. 
df[df['Name'].str.contains('United')]
# the set_index() function is used to set a column as the index of the dataframe. 
df2 = df.set_index('Name')
#filter() is used to only get some of the columns in the dataframe. axis is used to specify if you want to filter rows or columns. 0 is for rows, 1 is for columns.
df2.filter(items = ['Industry', 'Revenue (USD millions)'], axis = 1)
```

```python
# we can use the sort_values() function to sort the dataframe by a specific column. The ascending argument is used to specify if you want to sort in ascending or descending order.
# you can input a list of columns to sort by multiple columns. It sorts by the first item in the list, reading left to right. you can also specify if you want the other 
# items you're sorting by to be ascending or descending by inputting a list of booleans in the ascending argument.
df[df['Rank'] < 10].sort_values(by=['Revenue growth', 'Industry'], ascending=[False, True])
```

```python
group_by_frame = df.groupby('Industry')
group_by_frame.mean(numeric_only=True)
```

```python
df.groupby('Industry').count()
```

```python
df.groupby('Industry').min()
```

```python
#.agg() is used to aggregate the data in a column. You can use it to get the mean, max, min, count, and sum of a column. You can also use it to get the mean, max, min, count, and sum of multiple columns at once.
# .agg() needs to be passed a dictionary in order to work
# you can pass groupby() a list of columns
df.groupby('Employees').agg({'Rank':['mean','max','count','sum']})
```

```python
df.groupby(['Employees', 'Industry']).mean(numeric_only=True)
```
