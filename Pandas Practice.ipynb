```python
df = pd.read_excel(r'C:/Users/Miles/Downloads/Github files/readfiles/Customer Call List.xlsx')
df
```

```python
# its good to drop duplicates in a dataframe before doing anything else, unless of course you want to keep them for some reason.
df = df.drop_duplicates()
```

```python
# we can drop columns that we dont need in the dataframe using the drop() function. The columns argument is used to specify which columns to drop.
df = df.drop(columns = 'Not_Useful_Column')
```

```python
# .str specifies that we are using a string method on the column. .strip() removes whitespace by default. You can put everything you want to strip
# in one single string, and it will remove it
df['Last_Name'] = df['Last_Name'].str.strip('...123_,/')
```

```python
# By using .replace() and a little bit of regex, we can replace all non numeric or alphabetic values with empty space. we need to specify that we 
# are using regex.
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]','',regex=True)
# by using a lambda, we can turn all values from float to string
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
# Since we turned all the values into a string, we can use a lamda to add dashes between certain characters
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# this section of the code removes the na entries, and replaces them with blank space.
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
# by using the split() function, we can tell pandas to split the string at a certain value, how many splits we would like, and if we would like them in
# different columns. We MUST specify with n how many splits we are doing, as otherwise it will throw errors.
# We add the names of the new columns at the beginning in a list. 
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',',n=2, expand=True)
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes','Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes','Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a', '')
```

```python
# fillna() fills in blank values, for example, NaN
df.fillna('')
```

```python
# using a for loop, we can remove all of the customers who have set to do not contact. we do this by looping through the index, checking 
# if the specific value is equal to Y using .loc  
# we use the .drop() function with the inplace attribute to remove the whole row. 
for i in df.index:
    if df.loc[i, 'Do_Not_Contact'] == 'Y':
        df.drop(i, inplace=True)
df    
```

```python
for i in df.index:
    if df.loc[i, 'Phone_Number'] == '':
        df.drop(i, inplace=True)
df = df.reset_index(drop=True)
df
```
